## Explore FLAIR images
This notebook demonstrate the types of images in the dataset, and how to match images to metadata.

```python
import json
import os
from collections import defaultdict


dataset_dir = "../flair-data/"  # replace with the path to directory that you downloaded the dataset

with open(os.path.join(dataset_dir, "labels_and_metadata.json")) as f:
    metadata_list = json.load(f)


# Group images and metadata by user
image_metadata = {}
user_image_list = defaultdict(list)
for metadata in metadata_list:
    user_image_list[metadata["user_id"]].append(metadata["image_id"])
    image_metadata[metadata["image_id"]] = metadata
```

```python
from IPython.display import Image, display


image_dir = os.path.join(dataset_dir, "small_images")
# sort user ids by number of images each user has in descending order
user_ids = sorted(user_image_list, key=lambda k: -len(user_image_list[k]))


def print_metadata(metadata):
    print('-' * 100)
    print(f"User ID: {metadata['user_id']}, Image ID: {metadata['image_id']}")
    print(f"Labels: {metadata['labels']}")
    print(f"Fine-grained labels: {metadata['fine_grained_labels']}")


def display_user_images(user_id: str, max_to_display: int = 5):
    for image_id in user_image_list[user_id][:max_to_display]:
        print_metadata(image_metadata[image_id])
        display(Image(filename=os.path.join(image_dir, image_id + ".jpg")))
```

### Show 5 sample images for each of the 5 users with the greatest amount of data.
It is expected that `fine_grained_labels` contain `labels` as the subjects in an image might not be categorized into finer granularity.

```python
for user_id in user_ids[:5]:
    display_user_images(user_id)
```

### Show sample images for each of the 10 users with the least amount of data

```python
for user_id in user_ids[-10:]:
    display_user_images(user_id)
```
